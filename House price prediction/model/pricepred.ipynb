Now let's take a look at the 'total_sqft' column

```python
df2['total_sqft'].unique()
```

At first glance we see the values are in string. and there are variations like range values

```python
def is_float_possible(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df2[df2['total_sqft'].apply(is_float_possible)] # it's giving me all the rows that returns true for the function is applied
```

```python
df2[~df2['total_sqft'].apply(is_float_possible)].head(10) # it's giving me all the rows that returns False for the function is applied
```

We can see other than range values there are also some values like Sq.metre, perch etc

```python
def convert_into_float(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1]))/2
    else:
        try:
            return float(x)
        except:
            return x
```

```python
df3 = df2.copy()
```

```python
df3['total_sqft'] = df3['total_sqft'].apply(convert_into_float)
df3
```

```python
df3[~df3['total_sqft'].apply(is_float_possible)].shape
```

Other than range values there are more 46 different units values. we can just ignore them becase the number is so small caompare to 13246

```python
df4 = df3[df3['total_sqft'].apply(is_float_possible)]
df4
```

```python
df4['price_per_Sqft'] = df4['price']*100000/df4['total_sqft']
df4
```

Now let's remove outliers

```python
df5.shape
```

```python
df5 = df5[~(df5.total_sqft/df5.bedrooms<300)]
```

```python
df5.isna().sum()
```

```python
df5.shape
```

```python
df5['price_per_Sqft'].std()
```

```python
df5.groupby(['location'])['price_per_Sqft'].agg(['mean','std'])
```

```python
def clear_pps_outliars(x):
    df_out = pd.DataFrame()
    for key, subdf in x.groupby('location'):
        m = np.mean(subdf.price_per_Sqft)
        st = np.std(subdf.price_per_Sqft)
        reduce_df = subdf[(subdf.price_per_Sqft >= (m-st)) & (subdf.price_per_Sqft <= (m+st)) ]
        df_out = pd.concat([df_out, reduce_df], ignore_index=True)
    return df_out
```

```python
df6 = clear_pps_outliars(df5)
```

```python
df6.drop('size', axis=1, inplace=True)
df6
```

there is some outliars like in same location and square ft area for less bedroom the price is higher than higher bedrooms. So I want to remove those outliars.

```python
def remove_bedprice_inaquaty(x):
    rows_to_remove = []
    for _, group in x.groupby(['location','total_sqft']):
        group = group.sort_values(by='bedrooms')
        for i in range(len(group)-1):
            if group.iloc[i]['price_per_Sqft'] > group.iloc[i+1]['price_per_Sqft']:
                rows_to_remove.append(group.index[i])
    filtered_df = x.drop(rows_to_remove)
    return filtered_df
```

```python
df7 = remove_bedprice_inaquaty(df6)
df7.shape
```

```python
plt.hist(df7.price_per_Sqft, rwidth=0.8)
plt.show()
```

# Creating Model

Making dummy variables for location

```python
dummies = pd.get_dummies(df8.location).astype('int')
dummies
```

```python
df9 = pd.concat([df8,dummies.drop('other', axis=1)], axis=1)
df9
```

```python
df10 = df9.drop('location', axis=1)
df10
```

```python
X = df10.drop('price', axis=1)
X
```

```python
Y = df10.price
Y
```

```python
from sklearn.model_selection import train_test_split, ShuffleSplit, cross_val_score, GridSearchCV
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.tree import DecisionTreeRegressor
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=10)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
model.score(x_test,y_test)
```

```python
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
cvscore = cross_val_score(LinearRegression(), X,Y, cv=cv)
cvscore
```

Now I'll try to fit this dataset into various regressor model using gridsearch method

```python
model_params = {
    'linear_regression': {
        'model': LinearRegression(),
        'params' : {
        }  
    },
    'lasso': {
        'model': Lasso(),
        'params' : {
            'alpha': [1,2],
            'selection': ['random','cyclic']
        }
    },
    'decision_tree': {
        'model': DecisionTreeRegressor(),
        'params': {
            'criterion': ['mse','friedman_mse'],
            'splitter': ['best','random']   
        }
    }     
}
```

```python
scores = []
cvgrd = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
for i, j in model_params.items():
    clf = GridSearchCV(j['model'], j['params'], cv=cvgrd, return_train_score=False)
    clf.fit(X,Y)
    scores.append({
        'model': i,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
```

```python
pd.DataFrame(scores, columns=['model','best_score','best_params'])
```

So the winner is Linear regression. and I've already created a model using linear regression which gives me even better score 87%. So, I'm just going to use that model instead of making a new one

```python
def predict_price(location, sqft, bath, bedrooms):
    loc_index = np.where(X.columns == location)[0][0]
    x = np.zeros(len(X.columns))

    x[0] = sqft
    x[1] = bath
    x[2] = bedrooms
    if loc_index > 0:
        x[loc_index] = 1

    return model.predict([x])[0]
```
